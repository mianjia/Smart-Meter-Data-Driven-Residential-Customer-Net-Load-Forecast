```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import load traces of all available users 'data_hour.npy'
# data_hour should be size K*T, where K represents the total number of users, T represents total length of load traces
data_hour = np.load('data_hour.npy')
```

```python
from scipy.cluster import hierarchy
# tune threshold according to the dendrogram
threshold = 230
Z = hierarchy.linkage(data_hour, 'ward')
plt.figure(figsize=(40,15))
hierarchy.dendrogram(Z,color_threshold=230)
plt.axhline(y=threshold, color='black', linestyle='-')
plt.title('Dendrogram of All Users in Clifton Park Dataset',fontsize=30)
plt.savefig('dendrogram',bbox_inches='tight')
```

```python
# find corresponding customers of each cluster
clustering = AgglomerativeClustering(linkage="ward",n_clusters=11).fit(data_hour)
labels = clustering.labels_.tolist()
for i in range(11):
    print([j+1 for j, e in enumerate(labels) if e == i])
```
